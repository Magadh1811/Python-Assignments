### 1. List all flights scheduled for 1 Jan

```python
jan = df[(df["month"] == 1) & (df["day"] == 1)]
jan
```

### 2. Find flights that were not delayed more than 2 hours

```python
delay = df[df["dep_delay"] <= 120]
delay
```

### 3. Find all flights

#### a. Had an delay of 2 or more hours

```python
a = df[df["dep_delay"] >= 120]
a
```

#### b. Flew to Houston (IAH or HOU)

```python
airports = ["IAH", "HOU"]
b = df[df["dest"].isin(airports)]
b
```

#### c. Were operated by Unites, American or Delta

```python
flights = ["UA", "AA", "DL"]
c = df[df["carrier"].isin(flights)]
c
```

#### d. Departed in summer(March, April , May)

```python
d = df[(df["month"] >= 3) & (df["month"] <= 5)]
d
```

#### e. Arrived more than 2 hours late but didnâ€™t leave late

```python
e = df[(df["dep_delay"] <= 0) & (df["arr_delay"] >= 120)]
e
```

#### f. Departed between midnight and 6.00 am

```python
df["dep_time"] = pd.to_datetime(df["dep_time"], format = '%H%M', errors = 'coerce')
f = df[(df["dep_time"].dt.hour >= 0) & (df["dep_time"].dt.hour < 6)]
f
```

### 4. Sort flights to find most delayed flight

```python
sorted_flights = df.sort_values("arr_delay", ascending = False)
sorted_flights.iloc[0]
```

### 5. Which flights travelled longest?

```python
df["distance_covered"] = df["distance"] * 1.609344
sorted_flights = df.sort_values("distance_covered", ascending=False)
sorted_flights.iloc[0]
```

### 7. Convert dep_time and sched_dep_time to number of minute from midnight for further calculation

```python
df["dep_time"] = pd.to_datetime(df["dep_time"], format="%H%M", errors="coerce")
df["sched_dep_time"] = pd.to_datetime(df["sched_dep_time"], format="%H%M", errors="coerce")

df["dep_time_mins"] = df["dep_time"].dt.hour * 60 + df["dep_time"].dt.minute
df["sched_dep_time_mins"] = df["sched_dep_time"].dt.hour * 60 + df["sched_dep_time"].dt.minute
```
