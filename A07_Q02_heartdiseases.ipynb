Heart Diseases Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'D:\Msc DS\sem-2\PY assigments\A-07\heartdiseases.csv')
df
```

### 1. Load csv and set value -9 to NA

```python
df = df.replace(-9,pd.NA)
df
```

### 2. Processing will be focused on first 58 columns.
#### a. Take subset of first 58 columns

```python
df = df.iloc[:, 0:58]
df
```

### 3. Values 58 column defines how many times a patient had a heart attack
#### a. Since for building predictive model, letâ€™s focus on only 2 values: 0 means no attack, 1: possibility of attack. Therefore values > 1 should be reset to 1

```python
df['V58'] = np.where(df['V58'] > 1, 1 , df['V58'])
df
```

### 4. Since its a binary predictive model, we will process only numeric data
#### a. Check data types of each column

```python
df.dtypes
```

#### b. Drop columns which are having strings or objects

```python
df_2 = df.select_dtypes(include = np.number)
df_2.dtypes
```

### 5. Drop columns which have all null values
#### a. Check columns which have all null values // is.na, all() V5', 'V6', 'V45', 'V46', 'V47', 'V48', 'V49', 'V52', 'V53

```python
df.isna().all()
```

#### b. Drop the columns which have all null data

```python
df = df.dropna(axis = 1, how = 'all')
df
```

```python
# recheck after dropping NAN's
df.isna().all()
```

### 6. Find rows with missing values greater than 50%

```python
missing_values = df.isna()
missing_fraction = missing_values.sum(axis=1) / df.shape[1]
rows_with_missing = df[missing_fraction > 0.5]
print('rows with missing values greater than 50%:\n')
for row_index in rows_with_missing.index:
    print("Row", row_index)
```

### 7. Fill missing values with median value for all columns which are float or int
#### a. Print columns that have minimum one NaN value

```python
nan_1 = df.isna().any()
print(nan_1[nan_1].index)
```

#### b. For each value, Update Nan values with median value of that column Note: character vectors and factors are imputed with the mode. Numeric and integer vectors are imputed with the median.

```python
df = df.fillna(df.median())
df
```

#### c. Verify that no more Na values

```python
df.isna().sum()
```

### 8. Remove columns which have same value in all rows // since such column can not contribute to prediction

```python
nunique = df.nunique()
keep_cols_mask = (nunique > 1).values
df = df.loc[:, keep_cols_mask]
df
```
